```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('C:/Users/joker/Desktop/기계학습/lab3_test.csv')
setosa = df[df['Unnamed: 4'] == 'Iris-setosa']
versicolor = df[df['Unnamed: 4'] == 'Iris-versicolor']
virginica = df[df['Unnamed: 4'] == 'Iris-virginica']

plt.figure(figsize=(20, 10))

plt.subplot(1,2,1)
plt.scatter(setosa['sepal length'], setosa['sepal width'], s=255)
plt.scatter(versicolor['sepal length'], versicolor['sepal width'], s=255)
plt.scatter(virginica['sepal length'], virginica['sepal width'], s=255)
plt.legend(['Iris-setosa','Iris-versicolor','Iris-virginica'], fontsize=20)
plt.xlabel('sepal length', fontsize=25)
plt.ylabel('sepal width', fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('Figure 1', fontsize=35)

plt.subplot(1,2,2)
plt.scatter(setosa['petal length'], setosa['petal width'], s=255)
plt.scatter(versicolor['petal length'], versicolor['petal width'], s=255)
plt.scatter(virginica['petal length'], virginica['petal width'], s=255)
plt.legend(['Iris-setosa','Iris-versicolor','Iris-virginica'], fontsize=20)
plt.xlabel('petal length', fontsize=25)
plt.ylabel('petal width', fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('Figure 2', fontsize=35)

plt.tight_layout()
plt.show()
```
