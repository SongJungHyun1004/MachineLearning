```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

datas = pd.read_csv('C:/Users/joker/Desktop/기계학습/tic-tac-toe.csv', index_col=-1)
labels = datas.index
trn_datas, test_datas, trn_labels, test_labels = train_test_split(datas, labels, random_state=1)

# gini index 비교 후 속성 선택
attributions = datas.columns
gini_index = 0.5    #최대 지니 인덱스
selected_attr = ''
final_classifier = {} #최종 결정되는 분류기
ginis = []
for a in attributions:
    classifier = {}
    c = np.unique(trn_datas[a])
    std = []
    rest = []
    G = 0.5
    for i in c:
        std = trn_datas[a][trn_datas[a][:] == i]
        rest = trn_datas[a][trn_datas[a][:] != i]
        label = np.unique(trn_labels)
        GL = 0.0
        GR = 0.0
        std_majority = 0
        std_major = ''
        rest_majority = 0
        rest_major = ''
        for j in label:
            std_num = len(std[std.index == j])
            if std_num > std_majority:
                std_majority = std_num
                std_major = j
            pl = std_num / len(std)
            GL += pl * (1-pl)
            rest_num = len(rest[rest.index == j])
            if rest_num > rest_majority:
                rest_majority = rest_num
                rest_major = j
            pr = rest_num / len(rest)
            GR += pr * (1-pr)
        g = len(std)/len(trn_datas[a])*GL + len(rest)/len(trn_datas[a])*GR
        if g < G:
            G = g
            classifier[i] = std_major
            for r in np.unique(rest):
                classifier[r] = rest_major
    ginis.append(G)
    if G < gini_index:
        final_classifier = classifier
        gini_index = G
        selected_attr = a
        
df = pd.DataFrame(columns=attributions)
df.loc[0] = ginis
print('Selected attribution : ', selected_attr)
print('Minimum gini index : ', gini_index)
print('Tree : ', final_classifier)

#  분류기에 test data 넣어서 predict label 출력
predict_labels = []
for t in test_datas[selected_attr]:
    predict_labels.append(final_classifier[t])
    
test_labels = list(test_labels)
# test set으로 정확도(= (정답개수/test set 개수)*100 ) 측정
correct = 0
for i in range(len(predict_labels)):
    if predict_labels[i] == test_labels[i]:
        correct += 1
accuracy =  correct / len(test_labels) * 100
print('accuracy : ', accuracy)
df #속성 마다 지니인덱스 결과
```
